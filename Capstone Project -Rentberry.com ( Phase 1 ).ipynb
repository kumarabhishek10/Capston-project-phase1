```python
austin_url=[]
start = 0
end = 10
page_url = url[0]
driver.get(page_url)
full=driver.find_elements(By.XPATH,'/html/body/rb-core/div/rb-home-page/rb-promo-search/div/div/div/div/div[2]/input')
time.sleep(20)
```

```python
for i in range(start, end):
    for i in driver.find_elements(By.XPATH,'//div[@class="property-card-bottom-address rb-icon-before rb-icon--location"]/a'):
        austin_url.append(i.get_attribute('href'))
```

```python
austin_url
```

```python
house_name = []
location = []
rental_price = []
deposit = []

for i in austin_url:
    driver.get(i)
    time.sleep(30)
    
    h=driver.find_elements(By.XPATH,'//h1[@class="property-name"]')
    for element in h:
        try:
            house_name.append(element.text)
        except NoSuchElementException:
            house_name.append("-")
    
    
    l=driver.find_elements(By.XPATH,'//span[@class="property-data-address-name"]')
    for element in l:
        try:
            location.append(element.text)
        except NoSuchElementException:
            location.append("-")
    
    r=driver.find_elements(By.XPATH,'//div[@class="price-value-container"]')
    for element in r:
        try:
            rental_price.append(element.text)
        except NoSuchElementException:
            rental_price.append("-")
    
    d=driver.find_elements(By.XPATH,'/html/body/rb-core/div/rb-property/div/div/div/rb-property-main-info/section/div/div/rb-property-facilities/section/div[6]/span[2]')
    for element in d:
        try:
            deposit.append(element.text)
        except NoSuchElementException:
            deposit.append("-")
```
